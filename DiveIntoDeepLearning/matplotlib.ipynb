```python
# 同一位置多柱状图
# 对x轴进行计算
# 必须用数值型
# 先用数字标识 后面替换成国家
countries=["Norway","German","China","USA","Sweden"]
gold_medal=[16,12,9,8,8]
silver_medal=[8,10,4,10,5]
bronze_medal=[13,5,2,7,5]
x=np.arange(len(countries))
width=0.2

plt.xticks(x,labels=[countries[i] for i in x])
plt.bar(x-width,gold_medal,width=width,color="gold")
plt.bar(x,silver_medal,width=width,color="grey")
plt.bar(x+width,bronze_medal,width=width,color="saddlebrown")

for i in x:
    plt.text(x[i]-0.2,gold_medal[i],gold_medal[i],va="bottom",ha="center")
    plt.text(x[i], silver_medal[i], silver_medal[i], va="bottom", ha="center")
    plt.text(x[i]+0.2,bronze_medal[i],bronze_medal[i],va="bottom",ha="center")
```

```python
from cProfile import label


countries=["Norway","German","China","USA","Sweden"]
gold_medal=np.array([16,12,9,8,8])
silver_medal=np.array([8,10,4,10,5])
bronze_medal=np.array([13,5,2,7,5])
x=np.arange(len(countries))
width=0.5

plt.xticks(x,labels=[countries[i] for i in x])
plt.bar(x,gold_medal,width,label="gold",color="gold")
plt.bar(x,silver_medal,width,bottom=gold_medal,label="silver",color="grey")
plt.bar(x,bronze_medal,width,bottom=gold_medal+silver_medal,label="bronze",color="saddlebrown")

for i in x:
    plt.text(x[i],gold_medal[i]+silver_medal[i]+bronze_medal[i],gold_medal[i]+silver_medal[i]+bronze_medal[i],va="bottom",ha="center")
```

```python
movie=["1","2","3"]
real_day1=np.array([4053,2548,1543])
real_day2=np.array([7840,4013,2421])
real_day3=np.array([8080,3673,1342])
```

```python
y=np.arange(len(movie))
base1=np.zeros(len(movie))
base2=real_day1
base3=real_day1+real_day2
height=0.2
base1,base2,base3
```

```python
plt.barh(movie,real_day1,height=height)
plt.barh(movie,real_day2,left=base2,height=height)
plt.barh(movie,real_day3,left=base3,height=height)

sum_data=real_day1+real_day2+real_day3
for i in y:
    plt.text(sum_data[i],movie[i],sum_data[i],va="center",ha="left")
plt.xlim(0,sum_data.max()+2000)
```
