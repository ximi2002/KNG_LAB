```python
# 显示每条数据的值
times=np.arange(1990,2020).astype(np.str_)
income=np.random.randint(500,2000,size=len(times))
expenses=np.random.randint(500,2000,size=len(times))
plt.xticks(range(0,len(times),2),rotation=45)
plt.plot(times,income,label="income")
plt.plot(times,expenses,label="expenses")
plt.legend(loc="lower right")
for a,b in zip(times,income):
    plt.text(a,b,b)
for a,b in zip(times,expenses):
    plt.text(a,b,b)
```

```python
# 对坐标轴的操作
x=np.arange(-50,51)
y=x**2
plt.plot(x,y)
```

```python
# top,bottom,left,right
ax=plt.gca()

ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0.0))
ax.spines["bottom"].set_position(("data",0.0))

plt.plot(x,y)
```

```python
# top,bottom,left,right
ax=plt.gca()

ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
# 三种方式 data,outward,axes
# axes:0.0-1.0的值，整个轴上的比例
ax.spines["left"].set_position(("axes",0.5))
ax.spines["bottom"].set_position(("data",0.0))
# 设置坐标区间
plt.ylim(y.min(),y.max())
plt.plot(x,y)
```

```python
# 颜色
# "b" blue
# "g" green
# "r" red
# "c" cyan
# "m" magenta
# "y" yellow
# "k" black
# "w" white

# RGB 颜色对照表 16进制
# 颜色英文名称

# alpha 透明度 0-1

# linestyle 折线样式
# "-" 实线
# "--" 虚线
# "-." 点划线
# ":" 虚线

# 标记点
# "." 点标记
# "o" 圆圈标记
# "x" x标记
# "+" 加号标记
```
