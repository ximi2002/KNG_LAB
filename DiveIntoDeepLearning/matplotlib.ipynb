```python
# 对坐标轴的操作
x=np.arange(-50,51)
y=x**2
plt.plot(x,y)
```

```python
# top,bottom,left,right
ax=plt.gca()

ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position(("data",0.0))
ax.spines["bottom"].set_position(("data",0.0))

plt.plot(x,y)
```

```python
# top,bottom,left,right
ax=plt.gca()

ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
# 三种方式 data,outward,axes
# axes:0.0-1.0的值，整个轴上的比例
ax.spines["left"].set_position(("axes",0.5))
ax.spines["bottom"].set_position(("data",0.0))
# 设置坐标区间
plt.ylim(y.min(),y.max())
plt.plot(x,y)
```

```python
# 设置后再绘制
ax1=plt.subplot(211)
plt.title("ax1")
plt.plot(range(50,70))
ax2=plt.subplot(212)
plt.title("ax2")
plt.plot(np.arange(12)**2)

plt.tight_layout()
```

```python
# 先分割后设置

ax1=plt.subplot(211)
ax2=plt.subplot(212)

ax1.set_title("ax1")
ax1.plot(range(50,70))

ax2.set_title("ax2")
ax2.plot(np.arange(12)**2)

plt.tight_layout()
```

```python
# subplots()函数
fig,axes=plt.subplots(2,2)

ax1=axes[0][0]
x=np.arange(1,10)
ax1.plot(x,x*x)
ax1.set_title("square")

axes[0][1].plot(x,np.sqrt(x))
axes[0][1].set_title("square root")

axes[1][0].plot(x,np.exp(x))
axes[1][0].set_title("exp")

axes[1][1].plot(x,np.log10(x))
axes[1][1].set_title("log10")

plt.tight_layout()
```
