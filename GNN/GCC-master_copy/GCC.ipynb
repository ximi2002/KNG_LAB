```python
# 绘制预训练损失函数
import numpy as np
path_e2e="/home/suzhang/git/GNN/GCC-master/saved/Lossplot/loss_moco_False.txt"
path_moco="/home/suzhang/git/GNN/GCC-master/saved/Lossplot/loss_moco_True.txt"
path_fine_e2e_air="/home/suzhang/git/GNN/GCC-master/saved/Lossplot/loss_finetune_moco_False_usa_airport.txt"
path_fine_e2e_h_index="/home/suzhang/git/GNN/GCC-master/saved/Lossplot/loss_finetune_moco_False_h-index.txt"
```

```python
E2Eloss=np.loadtxt(path_e2e)
MoColoss=np.loadtxt(path_moco)
E2EAirloss=np.loadtxt(path_fine_e2e_air)
E2EHindexloss=np.loadtxt(path_fine_e2e_h_index)
```

```python
print(E2Eloss.shape,MoColoss.shape,E2EAirloss.shape,E2EHindexloss.shape)
```

```python
MoColoss[:,0]+=MoColoss[:,1]/MoColoss[:,2]
plt.plot(MoColoss[:,0],MoColoss[:,3])
```

```python
E2EAirloss[:,0]+=E2EAirloss[:,1]/E2EAirloss[:,2]
plt.plot(E2EAirloss[:,0],E2EAirloss[:,3])
```

```python
E2EHindexloss[:,0]+=E2EHindexloss[:,1]/E2EHindexloss[:,2]
plt.plot(E2EHindexloss[:,0],E2EHindexloss[:,3])
```

```python
from matplotlib import pyplot as plt
# 每个epoch的损失 
E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],0]+=E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],1]/E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],2]
plt.plot(E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],0],E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],3],label="loss1")
plt.plot(E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],0],E2Eloss[[i for i in range(0,E2Eloss.shape[0],9)],4],label="loss2")
leg=plt.legend()
```
