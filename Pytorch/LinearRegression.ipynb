### Linear Regression

```python
import torch
import numpy as np
from matplotlib import pyplot as plt 
```

#### 1.准备数据
y=3x+0.8

```python
x=torch.rand([500,1])
y_true=3*x+0.8
```

#### 2.计算y_predict

```python
w=torch.randn([1,1],requires_grad=True)
b=torch.randn([1,1],requires_grad=True,dtype=torch.float32)
y_predict=torch.matmul(x,w)+b
```

#### 3.计算loss

```python
loss=(y_true-y_predict).pow(2).mean()
```

#### 通过循环，反向传播，更新参数

```python
learning_rate=0.001

for i in range(10000):
    if w.grad is not None:
        w.grad.data.zero_()
    if b.grad is not None:
        b.grad.data.zero_()
        
    y_predict=torch.matmul(x,w)+b
    
    loss=(y_true-y_predict).pow(2).mean()
    
    loss.backward()
    w.data=w.data-learning_rate*w.grad
    b.data=b.data-learning_rate*b.grad
    if(i%100==0):
        print(str(i)+" w="+str(w.item())+" b="+str(b.item())+" loss="+str(loss.item()))
```

```python
plt.figure(figsize=(20,8))
plt.scatter(x.detach().numpy().reshape(-1),y_true.detach().numpy().reshape(-1),c="r")
y_predict=torch.matmul(x,w)+b
plt.scatter(x.detach().numpy().reshape(-1),y_predict.detach().numpy().reshape(-1))
plt.show()
```
