```python
from torch import nn
```

```python
class Lr(nn.Module):
    def __init__(self):
        super(Lr,self).__init__()
        self.linear=nn.Linear(1,1)
        
    def forward(self,x):
        out=self.linear(x)
        return out
```

```python
import torch
import numpy as np
from matplotlib import pyplot as plt 
```

定义数据

```python
x_true=torch.rand([500,1])
y_true=3*x_true+0.8
```

```python
model=Lr()
criterion=nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=1e-3)
for i in range(20000):
    y_predict=model(x_true)
    loss=criterion(y_true,y_predict)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (i%10==0):
        print(i,loss.data)
```

```python
model.eval()
```

```python
predict=model(x_true)
predict=predict.data.numpy()
plt.scatter(x_true.data.numpy(),y_true.data.numpy())
plt.plot(x_true.data.numpy(),predict)
```
